# Nested Sampling implementation of a simple microlensing model

In this notebook we apply the *Nested Sampling* (NS) algorithm (Skilling 2006) to the simplest model of a 
gravitational microlensing light curve, that of a single point lens illuminated by a point source (PSPL for short. A gravitational microlensing lightcurve consists of a series of one dimensional photometric flux measurements taken as successive times. Typically, microlensing surveys provide data consisisting 
of measurement times in Julian dates (assumed to have negligible uncertainty), flux measurements taken 
with a certain filter and measured in magnitudes, and the standard deviations of each of the flux
measurements assumed to be drawn from a Gaussian distribution. The mesured fluxes in the PSPL model 
are given by

$$F(t;t_0,u_0,t_E)=\Delta F\frac{A(t;t_0,u_0,t_E)-1}{A(u_0)-1}+F_{base}$$

where

$$A(u)=\frac{u^2+2}{u\sqrt{u^2+4}}$$

and

$$u(t;t_0,u_0,t_E)=\sqrt{u_0^2+\left(\frac{t-t_0}{t_E}\right)^2}$$

The model parameters are then $\boldsymbol{\theta}=\left\{\Delta F^{(k)},F_{base}^{(k)}, t_0,u_0,t_E\right\}$,
and the logarithm of the likelihood function is given by

$$\ln\mathcal{L}=-\frac{1}{2} \sum_{i=1}^N(2\pi\sigma_{Fi}^2) - \frac{1}{2}\sum_{i=1}^N\frac{\left[F_i - F(t;t_0,u_0,t_E)\right]^2}{\sigma_{Fi}^2}$$


The PSPL model (and any other more complex microlensing model) is non-linear and can 
exhibit multiple modes in the posterior distribution. While standard nonlinear optimization algorithms
and also sampling algorithms such as MCMC often get stuck in local extrema, this is (in principle) not the
case with NS. It is thus worthwhile to look into NS for microlensing modeling, especially considering the
fact that for more complex models of microlensing events, the situation only gets worse.


## Standard Nested Sampling algorithm

NS is a particular algorithm designed for Bayesian inference. Given the data $\mathbf{D}$ and a certain model
with parameters $\boldsymbol{\theta}$, the Bayes formula gives the posterior probability of the model parameters given the data:

$$p(\boldsymbol{\theta}\lvert\mathbf{D})=\frac{\mathcal{L}(\boldsymbol{\theta}\lvert\mathbf{D})\pi(
\boldsymbol{\theta})}{\mathcal{Z}}$$

where $\mathcal{L}$ is the likelihood function, $\pi$ is the prior distribution for the parameters, and
$\mathcal{Z}$ is the normalizing constant called the *Fully Marginalized Likelihood*, the
*The Bayesian Evidence*, or occasinally also *The Prior Predictive Distribution*. It is given by

$$\mathcal{Z}=p(\mathbf{D})=\int_{\boldsymbol{\theta}}\mathcal{L}(\boldsymbol{\theta}\lvert\mathbf{D})\pi(
\boldsymbol{\theta})d\boldsymbol{\theta}$$

It is a probability
distribution over the *data* before we have observed any data, marginalized over all of the model
parameters. Assuming we do have some data, it is just a number which acts as a normalization constant
for the posterior distribution $p(\boldsymbol{\theta}\lvert\mathbf{D})$.

In order to sample from the posterior distribution with a classic Markov Chain Monte Carlo (MCMC) 
algorithm, we only need to know the posterior distribution up to a constant so calculating $\mathcal{Z}$ 
explicitely is not necessary. However, if we are ever in a situation where we want to compare different
*models* for the same data, $\mathcal{Z}$ is usually the quantity we are interested in. 
Given two models $M_1$ and $M_2$, we can write down the ratio of the probabilities for the 
*models themselves* given the observed data, this quantity is known as the *Bayes factor*

$$\frac{p(M_1\lvert\mathbf{D})}{p(M_2\lvert\mathbf{D})}=\frac{p(M_1)}{p(M_2)}\frac{p(\mathbf{D}\lvert M_1)}
{p(\mathbf{D}\lvert M_2)}$$

The ratio of the two probabilities for the models is given by the *prior odds* $\frac{p(M_1)}{p(M_2)}$ 
times the ratio of the evidences for each model. Thus, if we ever want to compare different models in 
a Bayesian sense, we need to be able to calculate the evidence. Note that assigning probabilities to 
different models only makes sense in a Bayesian framework.

For any non-trivial model, $\mathcal{Z}$ is a complicated multi-dimensional integral which cannot be solved 
analytically. A simplest approach to calculating it is via Simple Monte Carlo integration, however that 
approach
usually fails for complex models. NS was designed specifically for calculating the evidence, although as
a by-product it also generates samples from the posterior just like any MCMC algorithm. The
key idea behind NS is that it is possible to cast the complicated multi-dimensional integral $\mathcal{Z}$ 
into a one-dimensional integral which is easily evaluated numerically.

We start by defining a function $X$ which maps the entire parameter space
to a number between 0 and 1, such that maximum of the
likelihood function  is mapped to $X=0$ and the minimum
is mapped to $X=1$, otherwise, for a given point $\boldsymbol{\theta}$ in the parameter space, we have

$$X(\mathcal{l})=\int_{\mathcal{L}(\boldsymbol{\theta})>\mathcal{l}}\pi(\boldsymbol{\theta})d
\boldsymbol{\theta}$$

That is, $X(\mathcal{l})$ is a number which tells us what fraction of the *total prior volume* has a 
likelihood value higher than some number $l$. $X(\mathcal{l})$ is a quantity similar to a cumulative
density function (CDF), except CDFs are only defined for one dimensional distributions.

Given $X$, the evidence integral simplifies to (Skilling 2006)

$$\mathcal{Z}=\int\mathcal{L}(X)dX$$

which is a simple one dimensional integral which can be easily evaluated numerically. 
The question then is how to compute a sufficient number of $X$ values to calculate the evidence integral. 
In a particular NS implementation, $X$ is computed for severals *levels* $l$, usually chosen such that 
$X(l)$ shrinks by some factor per each level.

<img src="nested_sampling.png" width="50%"><center>Figure 3 from Skilling (2006),
$L_1$, $L_2$ and $L_3$ are different levels.</center></img>

The standard version of the NS algorithm works by starting with a population of
$N$ *particles* (points in the parameter
space similar to *walkers* in ensemble MCMC methods) 
drawn from the prior. These are uniformly distributed
in terms of $X$ and are generally far away from points favoured by the data.
The particle with the lowest likelihood value, say $l_0$, is then recorded and replaced 
with a new sample from the *constrained* prior distribution, constrained such that $\mathcal{L}(\boldsymbol{\theta})>l_0$. The fraction of prior volume occupied by points with likelihood greater
than $l_0$ is then estimated to be 

$$X(l_0)\approx e^{-1/N}$$

To see why this is a sensible estimate, consider starting with $N=100$ particles and ranking them in terms
of their likelihood values. The fraction of prior volume above the one with the lowest likelihood is
approximately $99/100$ or

$$X(l_0)\approx\frac{N-1}{N}$$

since NS algorithms usually work with $\ln X$ instead of $X$, we have

$$\ln X=ln\left(1-\frac{1}{N}\right)\approx -\frac{1}{N}$$

Thus, at each iteration, the prior volume shrinks geometrically, starting from the whole prior volume, very
quickly we exponentially shrink down to the regions of largest likelihood values.

The algorithm is then as follows:

**Algorithm:**

* generate N points from the prior distribution $\pi(\boldsymbol{\theta})$
* find the point $\boldsymbol{\theta}'$ with the lowest likelihood value 
   $\mathcal{L}(\boldsymbol{\theta}')=l_0$. Record its approximate $X$ value as $X(l_0)\approx e^{-1/N}$
* replace the point $\boldsymbol{\theta}'$ with a point drawn from the prior $\pi(\boldsymbol{\theta})$
constrained to $\mathcal{L}(\boldsymbol{\theta})>l_0$
* repeat the last two steps for many levels $l_1$, $l_2$, $\dots$ until reaching some maximum level $l_M$

The key step is drawing the 
replacement point from the constrained prior distribution, there are several ways one can do this and it
is the key difference between various NS implementations. 
One way to do the third step is to copy the particle from with the next lowest value of likelihood after the one we discarded, and then run an MCMC chain which explores the prior distribution, except we reject all the 
points not satisfying the likelihood constraint, a method known as *rejection sampling*.

The discarded points and their $X$ values are the outputs of the algorithm and the one dimensional
evidence integral can be evaluated numerically using
standard quadrature methods as

$$\mathcal{Z}=\sum^M_{i=1}\mathcal{L}_is_i$$

where $N$ is the number of discarded particles
$\mathcal{L}_i=\mathcal{L}(X(l_i))$ and the weigths $s_i$ are given by
the trapezium rule:

$$s_i=\frac{1}{2}(X_{i-1}-X_{i+1})$$


Once the evidence $\mathcal{Z}$ is determined we can 
also obtain the samples from the posterior 
distribution by assigning the *posterior weights* $W_j$ to
each of the points with recorded $X$ values, as follows

$$W_j=\frac{\mathcal{L}_j s_j}{\mathcal{Z}}$$

these samples can then be used to construct posterior integrals of interest such as means, medians or 
full posterior histograms for each of the parameters (see Skilling 2006).

## Diffusive Nested Sampling

A different flavour of NS, called *Diffusive Nested Sampling* (DNS) was developed by Brewer et. al. (2011). The 
main differences between the standard version of NS and DNS are the following.

**Algorithm:**
+ Instead of drawing several particles from the prior, the algorithm starts by drawing one particle from
the prior, evolving it via MCMC for a set number of steps and storing the likelihood values of each of the points. One then determines
the $1-e^{-1}$ quantile of the samples and calls it $l_0$, thus creating a new prior volume $1/e$ times smaller. The points below are then discarded. The $X$ value of the $k$-th level is then approximately
$e^{-k}$
+ Then, instead of sampling the prior distribution constrained to $\mathcal{L}(\boldsymbol{\theta})>l_0$, the
algorithm samples instead a *mixture* (a weighted sum) of prior distributions constrained to each of the 
previously calculated levels. For example, at the third iteration, one would sample from the distribution

$$w_0\pi(\boldsymbol{\theta})\mathbb{1}(\mathcal{L}(\boldsymbol{\theta})>l_0)+
w_1\pi(\boldsymbol{\theta})\mathbb{1}(\mathcal{L}(\boldsymbol{\theta})>l_1)+
w_2\pi(\boldsymbol{\theta})\mathbb{1}(\mathcal{L}(\boldsymbol{\theta})>l_2)$$

   where $\mathbb{1}$ is the indicator function. This is nicely illustrated in Brewer et. al. (2011):

<img src="diffusive_nested_sampling.png" width="50%"><center>Figure 1 from Brewer et. al. (2011).</center></img>

+ Finally, the *mixture weights* are set to uniform and all of prior space is explored via MCMC in 
order to obtain more samples

The benefit of this approach is that the prior can be explored more freely because it can jump back to less
constrained distributions with some probability. 
This also means that it can better explore multi-modal distributions which is 
exactly what we want for complicated models in gravitational microlensing.

There are various choices for the mixture weights $w_j$ (where $j$ designates the current level),
 Brewer et. al. (2011) propose
 
 $$w_j\propto \exp (j/\lambda )$$
 
 where $\lambda$ is a scale length parameter which determines how often we sample from the less 
 constrained levels. After
 the desired number of levels is reached, we switch the mixture weights $w_j$ to uniform weights allowing the particle to
 explore the mixture at all levels.
 
The DNS algorithm which explores the mixture of contrained distributions can be more precisely stated
as MCMC sampling of the following target joint distribution over the parameter space and the levels:

$$p(\boldsymbol{\theta},j)=p(j)p(\boldsymbol{\theta}\lvert j)$$

where $p(j)=w_j$. It then follows that

$$p(\boldsymbol{\theta},j)=w_j \frac{\pi(\boldsymbol{\theta})\mathbb{1}(\mathcal{L}(\boldsymbol{\theta})>l_j)}
{\int \pi(\boldsymbol{\theta})\mathbb{1}(\mathcal{L}(\boldsymbol{\theta})>l_j)d\theta}
=\frac{w_j}{X(l_j)} \pi(\boldsymbol{\theta})\mathbb{1}(\mathcal{L}(\boldsymbol{\theta})>l_j)$$

In order to sample the above distribution with Metropolis Hastings MCMC, we need to know $X(l_j)$, 
fortunately, we know that it is by construction approximately $e^{-l}$.


Finally, DNS can also use multiple particles in order to better expolore highly correlated and multi-modal
distributions. When running with multiple particles DNS evolves them in parallel
and if one of the particles gets stuck in a mode, i.e. it is  at a distance of several times the scale length $\lambda$ away from the top level, it is deleted and more CPU time is available for the surviving particles (Brewer et. al. 2011).


This covers the basic description of the DNS algorithm, we now proceed to implement a model in the `DNest4`
`C++` code described in Brewer & Foreman-Mackey (2016). 

## DNest4 implementation

The `OPTIONS` file contains the specification of various input parameters required for running `DNest4`. The most important are the following:
+ Numer of particles per thread - effectively specifies the number of particles used in the code, one can use
more than one particles in DNS in order to obtain more samples and obtain a more precise estimate of the 
evidence
+ New level interval - specifies the nr. of MCMC steps taken by the Metropolis sampler before a new level
is created, 10000 steps is the default, depending on the problem a smaller number of steps might be sufficient
+ Maximum number of levels - specifies the maximum number of levels, if set to zero this number is determine
automatically, this may not always work but when it fails it will be evident in the diagnostic plot of 
posterior weigths vs $X$.
+ Backtracking scale length - $\lambda$ in the above description of the algorithm

```python
event = '2017/blg-0401' # '2017/0717'

data = pd.read_csv('../../microlensing_data/OGLE/' + event + '/phot.dat', 
                   names=['HJD', 'I_flux', 'I_flux_err', 'seeing_est', 'sky_level'],
                   delim_whitespace=True)

mag_to_flux = lambda m: 10**(-m/2.5)
magerr_to_fluxerr = lambda m, sigm: sigm*mag_to_flux(m)

data['HJD'] -= 2450000
data['I_flux_err'] = magerr_to_fluxerr(data['I_flux'].values, data['I_flux_err'].values)
data['I_flux'] = mag_to_flux(data['I_flux'].values)

# Scale flux units units 
data['I_flux'] *= 1e08
data['I_flux_err'] *= 1e08

# Save to .txt file, to be read by DNest4
np.savetxt('lightcurve.txt', data.values[:, :-2])
```

```python
# Compile DNest4
!make
```

```python
# Run DNest4 model on 4 threads
!./main -t 4
```

The run finished in ~15min running on 4 threads. Let's look at the diagnostic plots returned by the code:

<img src="diagnostic1.png" width="50%"><center></center></img>

The first plot shows the current level at each iteration. One can clearly see that the algorithm explores
ever higher levels, although the particle occasionally retreats to lower leveles. At the iteration of about
1500  the maximum level is reached, the weigths
$w_j$ are then all set to uniform and the MCMC chain is  free to move across all levels.

<img src="diagnostic2.png" width="50%"><center></center></img>

The first panel of the second plot shows the "Compression" of levels, that is, the natural logarithm
of $X(l_{j+1})/X(l_j)$, as a function of level number. By construction this quantity should be approximately 
equal to $\ln e^{-1}=-1$ which is indeed the case. The second panel shows the acceptence rate of the Metropolis
algorithm as a function of level number. Initially it starts out close to 1 because at low levels we are 
effectively sampling from the whole prior space and almost every Metropolis proposal is likely to get accepted.
As we move up in level the acceptance rate declines up to about $\sim0.2$ in the regions of highest likelihood.

<img src="diagnostic3.png" width="50%"><center></center></img>

The third plot is the most important one for judging a NS run. The first panel shows $\ln\mathcal{L}$
as a function of $\ln X$, it is to be read from right to left. As $\ln X$ decreases the likelihood
should increase. If there is a sort of saddle point in the curve where it becomes flat for a time then
then there is a *phase transition* in the problem (see Skilling 2006 for discussion), this is not the
case here because upon closer inspection the likelihood appears to be monotonically increasing with
decreasing prior volume fraction.

Finally, the second panel shows the posterior weigths $W_j$ as a function of $\ln X$, in a successfull run
there should be a clear peak in this plot without a hint of a second peak to the left. The presence of a 
second peak would indicate a phase transition in the problem (Brewer et. al. 2011).

```python
# Load the posterior samples
posterior_samples = np.loadtxt('posterior_sample.txt')

print(posterior_samples.shape)
print(np.percentile(posterior_samples[:, 4], [50])[0])
```

```python
import corner 
from IPython.display import display, Math

lbls = [r'$F_S$', r'$F_B$', r'$t_0$', r'$u_0$', 
                                     r'$t_E$', r'$\log_{10}\nu$', r'$u_K$']
labels = [r'F_S', r'F_B', r't_0', r'u_0$', 
                                     r'$t_E$', r'$\log_{10}\nu$', r'$u_K$']

ndim = 5

for i in range(ndim):
    mcmc = np.percentile(posterior_samples[:, i], [5, 50, 95])
    q = np.diff(mcmc)
    txt = "\mathrm{{{3}}} = {0:.3f}_{{-{1:.3f}}}^{{{2:.3f}}}"
    txt = txt.format(mcmc[1], q[0], q[1], labels[i])
    display(Math(txt))
    
fig = corner.corner(posterior_samples, labels=lbls[:-2])
```

```python
def pspl_flux(t, pars):
        """
        Evaluate a PSPL model at the input t values.


        Parameters
        ----------
        pars : list, array
            This should be a length-5 array or list containing the 
            parameter values (DeltaF, Fb, t0, u0, tE).
        t : numeric, list, array
            The time values.

        Returns
        -------
        F : array
            The computed flux values at each input t.
        """
            
        DeltaF, Fb, t0, u0, tE = pars

        u = np.sqrt(u0**2 + ((t - t0)/tE)**2)
        A = lambda u: (u**2 + 2)/(u*np.sqrt(u**2 + 4))

        return DeltaF*(A(u) - 1)/(A(u0) - 1) + Fb    

def plot_data(ax, t, F, sigF):
    ax.errorbar(t, F, sigF, fmt='.', color='black')
    ax.grid(True)
    ax.set_xlabel('$(t-\overline{t_0})$ [days]')
    ax.set_ylabel('$(F-\overline{F_b})/\overline{\Delta F}$')
    ax.set_title('OGLE-2017-BLG-0401');
    

def plot_model_and_data(ax, t, F, sigF, posterior):
    t0_median = np.percentile(posterior[:, 2], [50])[0]
    Fb_median = np.percentile(posterior[:, 1], [50])[0]
    DeltaF_median = np.percentile(posterior[:, 0], [50])[0]

    x = t - t0_median
    y = (F - Fb_median)/DeltaF_median

    # Plot data
    plot_data(ax, x, y, sigF/DeltaF_median)

    # Plot model in data space
    x_ = np.linspace(t[0], t[-1], 2000) 

    for pars in posterior[:128]: # only plot 128 samples
        y_ = (pspl_flux(x_, pars) - Fb_median)/DeltaF_median
        ax.plot(x_ - t0_median, y_,
                 marker='', linestyle='-', color='#3182bd', alpha=0.1, zorder=-10)

    ax.set_xlim(-100,100)
    
fig, ax = plt.subplots(figsize=(16,8))
plot_model_and_data(ax, data['HJD'].values, data['I_flux'].values, 
                    data['I_flux_err'].values, posterior_samples)
```

As can be seen from the traingle plot and plot of the model in data space, the DNS algorithm converged to an 
accurate solution to the problem.

# A different model for the data and a comparison between the two models

Since NS was constructed with the purpose of evaluating the evidence $\mathcal{Z}$, it would be interesting to
run it on a different model for the same data and compare the resulting Bayes factor in order to decide which
model is most likely. Instead of Gaussian noise in the likelihood function for the PSPL model, we can use
a Student's t distribution which is similar to a Gaussian except with wider tails. The likelihood function then
becomes

$$\mathcal{L}=\prod_{i=1}^N\frac{\Gamma\left(\frac{\nu+1}{2}\right)}{\Gamma\left(\frac{\nu}{2}\right)
\sigma_{fi}\sqrt{\pi\nu}}\left[1+\frac{1}{\nu}\frac{(F_i-F(t;t_0,u_0,t_E))^2}
{\sigma_{Fi}^2}\right]^{-\frac{\nu+1}{2}}$$

where $\nu$ is the *degree of freedom*, for $\nu>30$ the Student's t distribution is practically indistinguishable from a Gaussian. We choose to instead work with $\ln\nu$ as a parameter.
