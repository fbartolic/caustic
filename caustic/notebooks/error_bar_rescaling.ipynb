```python
random.seed(42)

data_path = '/home/star/fb90/data/OGLE_ews/2017/'

event = OGLEData(data_path + 'blg-0718')
```

```python
df = event.get_standardized_data()

F = df['I_flux'].values
sigF = df['I_flux_err'].values
sig_new = np.sqrt((2*sigF)**2 + (F*0.42/4)**2)

fig, ax = plt.subplots(figsize=(20,5))

ax.errorbar(df['HJD - 2450000'].values, F, sigF,fmt='.', color='black', label='Data', 
        ecolor='#686868' )
ax.errorbar(df['HJD - 2450000'].values, F, sig_new,fmt='.', color='black', label='Data', 
        ecolor='C0',alpha=0.5)
ax.grid(True)
```
